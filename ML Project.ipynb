```python
import json as _hex_json
_hex_pks.kernel_execution.scope_watcher.mark_dirty_scope_items(args=_hex_types.MarkDirtyScopeItemsArgs.from_dict({**_hex_json.loads("{\"dirty_scope_items\":[\"KMeans\",\"X\",\"inertias\",\"i\",\"kmeans\",\"labels\",\"data\",\"data\",\"plt\"]}")}), app_session_token=_hex_APP_SESSION_TOKEN, python_kernel_init_status=_hex_python_kernel_init_status, hex_timezone=_hex_kernel.variable_or_none("hex_timezone", scope_getter=lambda: globals()), interrupt_event=locals().get("_hex_interrupt_event"))

from sklearn.cluster import KMeans
X = data[['Median Household Income', 'Poverty Estimate', '2020-21 Per Capita Income', 
'Total  County Revenue', 'Total Local Appropriations', 'Total Local Appropriations  per ADM', 'Total no of crime acts committed', 
'Total no of crime acts committed per 1000 students',  'Total Students Enrolled', 'Total dropout count', 'Dropout Rate']]

# Normalize data
X = (X - X.mean()) / X.std()

# Determine optimal number of clusters
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
    inertias.append(kmeans.inertia_)
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Cluster data
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
labels = kmeans.labels_

# Add cluster labels to original data
data['Cluster'] = labels
data["Cluster"].value_counts()
```

```python
import json as _hex_json
_hex_pks.kernel_execution.scope_watcher.mark_dirty_scope_items(args=_hex_types.MarkDirtyScopeItemsArgs.from_dict({**_hex_json.loads("{\"dirty_scope_items\":[\"np\",\"plt\",\"sns\",\"StandardScaler\",\"train_test_split\",\"RandomForestRegressor\",\"mean_squared_error\",\"r2_score\",\"cross_val_score\",\"learning_curve\",\"columns\",\"data1\",\"X_train\",\"X_test\",\"y_train\",\"y_test\",\"scaler\",\"X_train_scaled\",\"X_test_scaled\",\"rf\",\"y_pred_scaled\",\"mse_scaled\",\"rmse_scaled\",\"r2_scaled\",\"data\"]}")}), app_session_token=_hex_APP_SESSION_TOKEN, python_kernel_init_status=_hex_python_kernel_init_status, hex_timezone=_hex_kernel.variable_or_none("hex_timezone", scope_getter=lambda: globals()), interrupt_event=locals().get("_hex_interrupt_event"))

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, learning_curve
from sklearn.metrics import mean_squared_error, r2_score

# select relevant columns
columns = ['Median Household Income', 'Poverty Estimate', 'Poverty Estimate(%)', 
           'Effective County Tax Rate', '2020-21 Per Capita Income', 'Total  County Revenue', 
           'Total Local Appropriations', 'Total Local Appropriations  per ADM', 
           'Total no of crime acts committed', 'Total Students Enrolled', 'Total dropout count']

data1 = data[columns]

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data1.drop('Total dropout count', axis=1), 
                                                    data1['Total dropout count'], 
                                                    test_size=0.2, random_state=42)

# standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# train random forest regression model on scaled data
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# predict on scaled test set
y_pred_scaled = rf.predict(X_test_scaled)

# evaluate model on scaled test set
mse_scaled = mean_squared_error(y_test, y_pred_scaled)
rmse_scaled = mse_scaled ** 0.5
r2_scaled = r2_score(y_test, y_pred_scaled)

print("Mean Squared Error (scaled): ", mse_scaled)
print("Root Mean Squared Error (scaled): ", rmse_scaled)
print("R-squared (scaled): ", r2_scaled)
```
